```python
import pandas as pd
iso = pd.read_csv("iso_3.csv", keep_default_na=False, na_values=['?'])
```

```python
iso
```

```python
iso['Code_continent'] = iso.Continent
```

```python
iso
```

```python
iso.Continent.replace('AF', 'Afrique', inplace=True)
iso.Continent.replace('AS', 'Asie', inplace=True)
iso.Continent.replace('EU', 'Europe', inplace=True)
iso.Continent.replace('OC', 'Océanie', inplace=True)
iso.Continent.replace('SA', 'Amerique du sud', inplace=True)
iso.Continent.replace('NA', 'Amerique du Nord', inplace=True)
iso.Continent.replace('AN', 'Antartique', inplace=True)
```
